```python
# - *  - utf-8 - *-
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.model_selection import train_test_split
import tensorflow as tf
from sklearn import metrics, cross_validation

warnings.filterwarnings('ignore')

%matplotlib inline

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 500)
```

## train data 나눈후 csv로 저장

```python
fashion_df = pd.read_csv('data/fashion-mnist_train.csv')
fashion_df.head()
```

```python
df_X = fashion_df.iloc[:, 1:]
df_y = fashion_df.iloc[:,0]
```

```python
df_X.tail()
```

```python
df_y.tail()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_X, df_y, test_size=0.3, random_state=0, shuffle = True)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
df_train = pd.concat([y_train,X_train], axis = 1)
```

```python
df_train.tail()
```

```python
df_test = pd.concat([y_test,X_test],axis=1)
```

```python
df_test.tail()
```

```python
df_train.to_csv('./data/fashion_train.csv', sep = ',', na_rep = 'NaN')
df_test.to_csv('./data/fashion_test.csv', sep = ',', na_rep = 'NaN')
```
